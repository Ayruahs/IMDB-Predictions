```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

filename = "movie_metadata.csv"
db = pd.read_csv(filename)

numericalList = []
for colname in db:
    if (type(db[colname][0]) != str):
        numericalList.append(colname)
 
numDataset = pd.DataFrame()
for numColName in numericalList:
    if numColName == 'imdb_score':
        continue
    else:
        numDataset[numColName] = db[numColName]
numDataset['imdb_score'] = db['imdb_score']
    
numDataset = numDataset.fillna(value=0, axis=1)
    
array = numDataset.values
array_standard = StandardScaler().fit_transform(array)
```

```python
i = 1
fig = plt.figure(figsize=[24,24])
for colName in numDataset:
    fig.add_subplot(4,4,i)
    plt.xlabel(colName)
    plt.ylabel('IMDB score')
    plt.scatter(db[colName], db['imdb_score'], c='c', alpha=0.4)
    i += 1
    #plt.scatter(array_standard[colName], array_standard[:,-1], c='blue', alpha=0.4)
plt.show()
plt.close()
```

```python
seed = 7
n = 10

# preparing models
models = [('RFR', RandomForestRegressor()), ('ETR', ExtraTreesRegressor()),
          ('GBR', GradientBoostingRegressor()), ('BR', BaggingRegressor()), 
          ('ABR', AdaBoostRegressor())]

#evaluate each model
results = []
names = []



for name, model in models:
    kfold = model_selection.KFold(n_splits=n, shuffle=True, random_state=seed)
    cv_results = model_selection.cross_val_score(model, array_standard[:,0:-1], array_standard[:,-1], cv=kfold)
    
    results.append(cv_results)
    names.append(name)
    
    print('%s: %.2f%% (%f)' %(name, cv_results.mean()*100, cv_results.std()))
    
'''array_standard = StandardScaler().fit_transform(array)
pca = PCA(n_components=len(numericalList))
array_16d = pca.fit_transform(array_standard)

plt.scatter(array_16d[:,0], array_16d[:,-1])
plt.show()'''
```
